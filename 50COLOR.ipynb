```python
import cv2
import os
import time

# Configuración
NUM_FOTOS = 50
CARPETA_BASE = 'dataset'

# Crear la carpeta principal si no existe
if not os.path.exists(CARPETA_BASE):
    os.makedirs(CARPETA_BASE)

# Cambia el 0 por 1 si tu webcam USB no es la principal de la laptop
cap = cv2.VideoCapture(1)

color = input(": Introduce la inicial del color a entrenar (R, V, A): ").strip().upper()
ruta_color = os.path.join(CARPETA_BASE, color)

if not os.path.exists(ruta_color):
    os.makedirs(ruta_color)

print(f"\n📸 Preparado para tomar {NUM_FOTOS} fotos de color {color}.")
print("Tienes 3 segundos para posicionar la pieza...")
time.sleep(3)

for i in range(NUM_FOTOS):
    ret, frame = cap.read()
    if ret:
        # Voltear horizontalmente para efecto espejo
        frame = cv2.flip(frame, 1)

        # DEFINIR ZONA DE INTERÉS (ROI):
        # Ajusta estos valores para que coincidan con el recuadro de tu banda
        y1, y2, x1, x2 = 200, 280, 280, 360
        zona_analisis = frame[y1:y2, x1:x2]

        # Dibujar recuadro guía en la pantalla
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.imshow("Recoleccion de Datos - WARRY", frame)

        # Guardar imagen
        nombre_archivo = os.path.join(ruta_color, f"img_{i}_{int(time.time())}.jpg")
        cv2.imwrite(nombre_archivo, zona_analisis)

        print(f"✅ Foto {i+1}/{NUM_FOTOS} guardada en {ruta_color}")

        # Pequeña pausa para que muevas ligeramente la pieza o cambies iluminación
        time.sleep(0.2)

        # Presiona 'q' para cancelar la recolección si es necesario
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
print("\n Recolección finalizada. ¡Dataset listo para entrenar!")
```
