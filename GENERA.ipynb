```python
import cv2
import numpy as np
import os

# 1. Crear las carpetas si no existen
carpetas = ['dataset/R', 'dataset/V', 'dataset/A']
for c in carpetas:
    os.makedirs(c, exist_ok=True)

print("Generando 150 fotografías .jpg físicas para tu IA...")

# 2. Generar 50 fotos por cada color simulando el lente de la cámara
for i in range(50):
    # Generar ruido visual simulado
    ruido = np.random.randint(0, 40, (80, 80, 3), dtype=np.uint8)

    # FOTO ROJA (Formato OpenCV BGR)
    rojo = np.full((80, 80, 3), (20, 20, 200), dtype=np.uint8) + ruido
    cv2.imwrite(f'dataset/R/foto_R_{i}.jpg', np.clip(rojo, 0, 255))

    # FOTO VERDE (Formato OpenCV BGR)
    verde = np.full((80, 80, 3), (20, 200, 20), dtype=np.uint8) + ruido
    cv2.imwrite(f'dataset/V/foto_V_{i}.jpg', np.clip(verde, 0, 255))

    # FOTO AZUL (Formato OpenCV BGR)
    azul = np.full((80, 80, 3), (200, 20, 20), dtype=np.uint8) + ruido
    cv2.imwrite(f'dataset/A/foto_A_{i}.jpg', np.clip(azul, 0, 255))

print("✅ ¡Éxito! Revisa la carpeta 'dataset', tus fotos reales ya están ahí listas para entrenar.")
```
